```python
import pandas as pd
import numpy as np
import warnings
import ast

```

```python
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("tmdb_5000_movies.csv")
df
```

```python
cred=pd.read_csv("tmdb_5000_credits.csv")
cred
```

```python
# generes id_x keywords overview production_companies title cast crew
```

```python
movies=df.merge(cred,on='title')
movies
```

```python
final_df=movies[["id","title","overview","genres","cast","crew",'keywords']]
final_df
```

```python
final_df.isnull().sum()
final_df.dropna(inplace=True)
final_df.isnull().sum()
```

```python
final_df["genres"][0]
```

```python
def change(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i["name"])
    return l
```

```python
final_df['genres']=final_df["genres"].apply(change)
final_df["genres"]
final_df['keywords']=final_df["keywords"].apply(change)
```

```python
final_df['crew']
print(final_df["cast"])
```

```python
def change1(text):
    l=[]
    count=0
    for i in ast.literal_eval(text):
        if count<3:
             l.append(i["name"])
             count+=1
        else:
             break
    return l
```

```python
def change2(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['department']=="Directing":
             l.append(i["name"])
             break
    return l
```

<!-- Here editing the crew cast and genres from dict to list-->

```python
final_df["cast"]=final_df["cast"].apply(change1)
```

```python
final_df['crew']=final_df['crew'].apply(change2)
final_df['crew']
```

```python
final_df
```

```python
final_df['overview']=final_df['overview'].apply(lambda x:x.split())
```

```python
final_df['crew']=final_df['crew'].apply(lambda x:[ i.replace(" ","") for i in x])
final_df['cast']=final_df['cast'].apply(lambda x:[ i.replace(" ","") for i in x])
```

```python
final_df['overview']
```

```python
final_df['lable']=final_df['overview']+ final_df["genres"]+ final_df['cast'] +final_df['keywords']+ final_df["crew"]
final_df["lable"]=final_df["lable"].apply(lambda x:" ".join(x))
del final_df['crew']


```

```python
del final_df['cast']
del final_df['overview']
del final_df['keywords']
```

```python
final_df
del final_df['genres']
```

```python
final_df
```

```python
final_df['lable'][0]
final_df['lable']=final_df['lable'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_df['lable']).toarray()
cv.get_feature_names()
```

<!-- Stemming the words or feature which are same like happen,happened -->

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    l=[]
    for i in text:
        l.append(ps.stem(i))
    return "".join(l)
```

```python
a=["actually"]
stem(a)
```

```python
final_df['lable']=final_df['lable'].apply(stem)
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
sorted(list(enumerate(similarity[0])),reverse =True,key=lambda x:x[1])
```

```python
def predict(movie):
    movie_idx=final_df[final_df['title']==movie].index[0]
    distance=similarity[movie_idx]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(final_df.iloc[i[0]].title)
```

```python
predict("Batman Begins")
```
